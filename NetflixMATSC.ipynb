## **<u>Project Summary</u>**

This project focuses on performing a comprehensive **Exploratory Data Analysis (EDA)** of Netflix’s catalog of movies and TV shows to uncover meaningful insights about the platform’s content strategy, audience preferences, and business direction.  
The dataset used contains **7,787 records and 12 columns**, providing information such as title, type (Movie or TV Show), director, cast, country, date added, release year, rating, duration, genres (listed_in), and description. The data was sourced from **Flixable**, a third-party Netflix search engine, and represents titles available on Netflix up to 2019.

In addition to EDA, this project serves as the foundation for a **machine learning (ML) task using K-Means clustering**, where the goal is to group similar titles based on features such as genre, rating, duration, and release period.  
The insights derived from EDA help define relevant features, guide preprocessing steps, and ensure the data is well-prepared for clustering analysis.

---

### **1. Data Cleaning and Preprocessing**

The initial step involved inspecting and cleaning the dataset. Columns such as `director`, `cast`, `country`, and `rating` contained missing values, which were replaced with `"Unknown"` to retain records without losing data.  
The `date_added` column was converted into a proper datetime format, and two new columns — `year_added` and `month_added` — were derived for trend analysis.  
The `duration` column was further split into `duration_minutes` (for movies) and `seasons` (for TV shows) to enable quantitative analysis across both types of content.  
The dataset was found to be free of duplicates, with minimal data inconsistencies, making it ready for both visual exploration and machine learning applications.

In 30.68%, 9.22%, 6.51%, 0.13%, and 0.09% of their respective features, director, cast, country, date_added, and rating had null values. These null values have been replaced by director Unavailable, Cast Unavailability, and Country Unavailable, respectively, because there are several null values for features like director, cast, and country. Due to the extremely low quantity of null values for features like date_added and rating, we eliminated those null values.

The interquartile range is successfully used to treat outliers from the variable release year. Before eliminating the feature date added, it was converted to datetime and used to generate additional features like year added, month added, and day added. Geners is the new name assigned to the listed in feature. The feature release year data type is changed from float64 to int64 because year cannot be a float.

#### **[3] <u>Handling Outliers</u>**

```python
## Checking for outliers in numerical variables using boxplot

# Importing library for checking normality in distribution
from scipy.stats import norm

# Plotting Box and Distribution plot using for loop
for var in numerical_variables:
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    ax=sns.boxplot(data=df[var])
    ax.set_title(f'{var}')
    ax.set_ylabel(var)

    plt.subplot(1,2,2)
    ax=sns.distplot(df[var], fit=norm)
    ax.set_title(f'skewness of {var} : {df[var].skew()}')
    ax.set_xlabel(var)
    print('__'*45)
    plt.show()
    print('__'*45)
```

- Variable release_year have outliers.

```python
## Using Inter Quantile Range for removing outliers from numerical variables

# Capping dataset
for i in numerical_variables:
    # Findling IQR
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    
    # Defining upper and lower limit
    lower_limit =df[i].quantile(0.25)-1.5*IQR
    print("Lower Limit:", lower_limit)
    upper_limit =df[i].quantile(0.75)+1.5*IQR
    print("Upper Limit:",upper_limit)
    # Applying lower and upper limit to each variables
    df.loc[(df[i] > upper_limit),i] = upper_limit
    df.loc[(df[i] < lower_limit),i] = lower_limit
```

```python
## Using boxplot checking outliers in features after removed using IQR

for var in numerical_variables:
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    ax=sns.boxplot(data=df[var])
    ax.set_title(f'{var}')
    ax.set_ylabel(var)

    plt.subplot(1,2,2)
    ax=sns.distplot(df[var], fit=norm)
    ax.set_title(f'skewness of {var} : {df[var].skew()}')
    ax.set_xlabel(var)
    print('__'*45)
    plt.show()
    print('__'*45)
```

- Outliers from variable release_year are successfully treated using the interquartile range.

```python
# Defining a function to Notate the percent count of each value on the bars
def annot_percent(axes):
    '''Takes axes as input and labels the percent count of each bar in a countplot'''
    for p in plot.patches:
        total = sum(p.get_height() for p in plot.patches)/100
        percent = round((p.get_height()/total),2)
        x = p.get_x() + p.get_width()/2
        y = p.get_height()
        plot.annotate(f'{percent}%', (x, y), ha='center', va='bottom')
```

#### 6. Rephrase Text

```python
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')

# Select text columns
text_columns = data.select_dtypes(include=['object']).columns
print("Text columns found:", list(text_columns))

# Simple function to rephrase text
def simple_rephrase(text):
    try:
        if isinstance(text, str):
            # Split into sentences
            sentences = sent_tokenize(text)
            rephrased_sentences = []
            for sent in sentences:
                # Remove duplicate words like "very very" → "very"
                sent = re.sub(r'\b(\w+)( \1\b)+', r'\1', sent)
                # Remove extra spaces
                sent = re.sub(r'\s+', ' ', sent)
                # Optional: Fix common patterns like "it's about" → "the story is about"
                sent = sent.replace("it's about", "the story is about")
                sent = sent.replace("its about", "the story is about")
                rephrased_sentences.append(sent.strip())
            return " ".join(rephrased_sentences)
        return text
    except Exception:
        return text

# Apply rephrasing to each text column
for col in text_columns:
    data[col] = data[col].apply(simple_rephrase)

print("Text rephrasing (basic cleanup) completed successfully!")
```

#### 8. Text Normalization

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download required NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Lemmatizer initialization
lemmatizer = WordNetLemmatizer()

# Function to get POS (part of speech) tag for better lemmatization
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# Function to normalize text (works on tokenized lists)
def normalize_text(tokens):
    try:
        if isinstance(tokens, list):
            lemmatized = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in tokens]
            return lemmatized
        elif isinstance(tokens, str):
            # If not tokenized, split and normalize
            words = tokens.split()
            lemmatized = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
            return lemmatized
        return tokens
    except Exception:
        return tokens

# Apply normalization to the tokenized text columns
text_cols_for_normalization = ['title_tokens', 'description_tokens']

for col in text_cols_for_normalization:
    data[col + '_normalized'] = data[col].apply(normalize_text)

print("Text normalization (lemmatization) completed successfully!")

# Example
print("\n Example:")
print("Before:", data['description_tokens'].iloc[0])
print("After Normalization:", data['description_tokens_normalized'].iloc[0])
```

##### Which text normalization technique have you used and why?

For text normalization, I used the Lemmatization technique provided by NLTK’s WordNetLemmatizer.
Lemmatization reduces each word to its base or dictionary form (known as the lemma) while considering the part of speech (POS) of the word.
For example, words like “playing”, “played”, and “plays” are all converted to “play”, and “children” becomes “child”.

I chose lemmatization over stemming because it is more accurate and context-aware — it doesn’t just cut off word endings but ensures that each word is linguistically correct.
This technique helps in reducing redundant variations of the same word and makes the text data more consistent and meaningful for machine learning models like clustering or sentiment analysis.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy.stats import skew, kurtosis


# Reduce Feature Correlation using PCA


print("Performing PCA to minimize correlated TF-IDF features...")

# Scale the data before PCA
scaler = StandardScaler(with_mean=False)  # with_mean=False for sparse data like TF-IDF
tfidf_scaled = scaler.fit_transform(tfidf_df)

# Apply PCA to reduce dimensionality (retain 95% variance)
pca = PCA(n_components=0.95, random_state=42)
tfidf_pca = pca.fit_transform(tfidf_scaled)

print(f"Reduced from {tfidf_df.shape[1]} → {tfidf_pca.shape[1]} principal components.")

# Convert PCA components back into a DataFrame
pca_df = pd.DataFrame(tfidf_pca, columns=[f'PC{i+1}' for i in range(tfidf_pca.shape[1])])

# Merge PCA-reduced features with original Netflix dataset
netflix_reduced_df = pd.concat([data.reset_index(drop=True), pca_df], axis=1)


#  Check Correlation Matrix (Post-PCA)

corr_matrix = pd.DataFrame(pca_df).corr()
high_corr = np.sum(np.abs(corr_matrix.values[np.triu_indices_from(corr_matrix, k=1)]) > 0.8)

print(f"Correlation check done — {high_corr} highly correlated pairs remaining (should be near 0).")


#  Create New Engineered Features


print("Creating new engineered features...")

# 🎬 Text-based Features
netflix_reduced_df['Description_Length'] = data['description'].str.split().apply(len)
netflix_reduced_df['Unique_Words'] = data['description'].apply(lambda x: len(set(str(x).split())))
netflix_reduced_df['Lexical_Richness'] = netflix_reduced_df['Unique_Words'] / netflix_reduced_df['Description_Length']

# Temporal Features
if 'release_year' in data.columns:
    netflix_reduced_df['Is_Recent'] = data['release_year'].apply(lambda x: 1 if x >= 2018 else 0)
    netflix_reduced_df['Title_Age'] = 2025 - data['release_year']

# Category/Genre Features
if 'listed_in' in data.columns:
    netflix_reduced_df['Num_Genres'] = data['listed_in'].apply(lambda x: len(str(x).split(',')))

# Statistical Features from PCA Components
netflix_reduced_df['PCA_Mean'] = pca_df.mean(axis=1)
netflix_reduced_df['PCA_Std'] = pca_df.std(axis=1)
netflix_reduced_df['PCA_Skew'] = pca_df.apply(lambda x: skew(x), axis=1)
netflix_reduced_df['PCA_Kurtosis'] = pca_df.apply(lambda x: kurtosis(x), axis=1)

# ------------------------------------------------
# 4️⃣ Final Feature Overview

print("\nFeature Manipulation Completed!\n")
print(f"Final dataset shape: {netflix_reduced_df.shape}")
print("\nNewly added features:")
print([
    'Description_Length', 'Unique_Words', 'Lexical_Richness',
    'Is_Recent', 'Title_Age', 'Num_Genres',
    'PCA_Mean', 'PCA_Std', 'PCA_Skew', 'PCA_Kurtosis'
])


#  Save for Next ML Steps

netflix_reduced_df.to_csv("Netflix_Feature_Engineered.csv", index=False)
print("\n Saved 'Netflix_Feature_Engineered.csv' successfully!")
```

```python
# FEATURE SELECTION TO AVOID OVERFITTING

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold, SelectKBest, mutual_info_classif, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score



# 1. Separate Features and Target

target_col = 'Cluster'  # adjust if needed
X = data.drop(columns=[target_col], errors='ignore')
y = data[target_col]

# Drop non-numeric features
X = X.select_dtypes(include=[np.number])

print("Initial shape of X:", X.shape)


# 2. Remove Near-Zero Variance Features

var_thresh = VarianceThreshold(threshold=0.01)
X_var = X.loc[:, var_thresh.fit(X).get_support()]
print("After variance threshold:", X_var.shape)


# 3. Remove Highly Correlated Features

corr_matrix = X_var.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_features = [col for col in upper.columns if any(upper[col] > 0.85)]
X_corr = X_var.drop(columns=high_corr_features)
print("Removed correlated features:", len(high_corr_features))
print("After correlation filter:", X_corr.shape)


# 4. Scale Features

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_corr), columns=X_corr.columns)


# 5. Mutual Information Selection

mi_selector = SelectKBest(score_func=mutual_info_classif, k=100)
mi_selector.fit(X_scaled, y)
mi_features = X_scaled.columns[mi_selector.get_support()]
X_mi = X_scaled[mi_features]
print("After Mutual Information Selection:", X_mi.shape)


# 6. Model-Based Recursive Feature Elimination (RFE)

log_reg = LogisticRegression(max_iter=1000, random_state=42)
rfe_selector = RFE(log_reg, n_features_to_select=50)
rfe_selector.fit(X_mi, y)

rfe_features = X_mi.columns[rfe_selector.get_support()]
X_rfe = X_mi[rfe_features]
print("After RFE Selection:", X_rfe.shape)


# 7. Stability Check Using Random Forest

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_rfe, y)
importances = pd.Series(rf.feature_importances_, index=X_rfe.columns)
selected_features = importances.sort_values(ascending=False).head(30).index

X_final = X_rfe[selected_features]
print("Final selected feature count:", X_final.shape[1])


# 8. Evaluate for Overfitting Risk

X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)

rf_eval = RandomForestClassifier(n_estimators=100, random_state=42)
rf_eval.fit(X_train, y_train)

train_acc = accuracy_score(y_train, rf_eval.predict(X_train))
test_acc = accuracy_score(y_test, rf_eval.predict(X_test))

print("\nModel evaluation to check overfitting:")
print("Train Accuracy:", round(train_acc, 4))
print("Test Accuracy :", round(test_acc, 4))
print("Generalization Gap:", round(train_acc - test_acc, 4))


# 9. Save Final Selected Features

final_data = pd.concat([X_final, y.reset_index(drop=True)], axis=1)
final_data.to_csv("Netflix_Feature_Selected_Final.csv", index=False)
print("\nFeature selection completed and saved as 'Netflix_Feature_Selected_Final.csv'")
```

#### 2. Cross- Validation & Hyperparameter Tuning

```python
# RANDOM FOREST - CROSS VALIDATION & TUNING

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# ------------------------------------------------
# Define the EXACT same parameter grid
# ------------------------------------------------
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# ------------------------------------------------
# Speed Boost: Added 'max_samples' and 'n_jobs' inside the RF
# ------------------------------------------------
rf = RandomForestClassifier(
    random_state=42, 
    n_jobs=-1,         # Internal parallelization
    max_samples=0.5    # Each tree only sees 50% of data (huge speedup, often same accuracy)
)

# ------------------------------------------------
# GridSearchCV - Reduced to 3-Fold for speed
# ------------------------------------------------
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,              # <--- Changed from 5 to 3 (saves 40% of the time)
    n_jobs=-1,         # Use all CPU cores
    verbose=1,         # Keeps you updated on progress
    scoring='accuracy'
)

# Fit using your encoded data
grid_search.fit(X_train, y_train_encoded)

# ------------------------------------------------
# Best parameters and Evaluation
# ------------------------------------------------
print("\nBest Parameters:", grid_search.best_params_)
print("Best CV Score:", round(grid_search.best_score_, 4))

best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test)

print("\nTuned Random Forest Accuracy:", round(accuracy_score(y_test_encoded, y_pred_best), 4))
print("\nClassification Report:\n", classification_report(y_test_encoded, y_pred_best))
```

```python
# MODEL 3 - XGBOOST CLASSIFIER


import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
X_train = pd.read_csv("Netflix_X_train_balanced.csv")
y_train = pd.read_csv("Netflix_y_train_balanced.csv").squeeze()
X_test = pd.read_csv("Netflix_X_test.csv")
y_test = pd.read_csv("Netflix_y_test.csv").squeeze()

# Encode labels
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)


# Fit the Algorithm

xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='mlogloss'
)
xgb_model.fit(X_train, y_train_encoded)


# Predict on the Model

y_pred = xgb_model.predict(X_test)


# Evaluate Model Performance

accuracy = accuracy_score(y_test_encoded, y_pred)
print("XGBoost Accuracy:", round(accuracy, 4))
print("\nClassification Report:\n", classification_report(y_test_encoded, y_pred))

# Confusion Matrix Visualization
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test_encoded, y_pred), annot=True, fmt='d', cmap='Purples')
plt.title('XGBoost Classifier - Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

The third model used is the XGBoost Classifier, which is a high-performance implementation of Gradient Boosting.
It builds multiple weak learners (decision trees) sequentially — each new tree corrects the errors made by the previous ones.
Unlike Random Forest (which builds trees independently), XGBoost focuses on learning from mistakes, optimizing for both bias and variance.

I chose XGBoost because:

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# 1. Define a smaller, high-impact grid (12 combinations)
param_grid = {
    'n_estimators': [100, 200],      # Fewer trees to start
    'learning_rate': [0.05, 0.1],     # Standard high-performance rates
    'max_depth': [4, 6, 8],          # Depth 8 is the limit for speed
}

# 2. Use 'tree_method="hist"' for a massive speed boost
xgb = XGBClassifier(
    random_state=42, 
    eval_metric='mlogloss', 
    tree_method='hist',   # <--- CRITICAL for speed
    n_jobs=-1             # Use all cores within the model too
)

# 3. Use 3-Fold CV instead of 5
grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=3,                 # <--- Faster than 5-fold
    scoring='accuracy',
    n_jobs=-1,            # Parallel processing
    verbose=1
)

# 4. Fit the model
grid_search.fit(X_train, y_train_encoded)

print("Best Parameters:", grid_search.best_params_)
print("Best Score:", round(grid_search.best_score_, 4))
```
